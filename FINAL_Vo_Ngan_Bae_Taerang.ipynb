```python
import turtle 
import time 

scr=turtle.getscreen()
scr.title("Flag")
scr.bgcolor("#FFFDD0")

t=turtle.Turtle()
t.speed(100)
t.penup()

#French Flag 
## stripes
def draw_rectangle(x,y,height,width,color):
    t.goto(x,y)
    t.pendown()
    t.color(color)
    t.begin_fill()
    
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    
    t.end_fill()
    t.penup()

    #stripes
def draw_stripes():
    x = -152
    y = 108
    height = 200
    width = height*1.9

# red and white stripes (total 13 stripes in flag)
    stripe_ht = height/13
    stripe_wdt = width

    # draw 6 red and 7 white strips
    for stripe in range(0,6):
        draw_rectangle(x,y, stripe_ht, stripe_wdt, '#B31942')
        # decrease value of y by stripe height
        y = y - stripe_ht
        draw_rectangle(x,y, stripe_ht, stripe_wdt, '#FFFFFF')
        y = y - stripe_ht

    # bottom/last red stripe
    draw_rectangle(x,y, stripe_ht, stripe_wdt, '#B22234')
    y = y - stripe_ht
    

## canton
def draw_canton():
    color = "#0A3161"
    height = 200
    width = height*1.9
    c_width = width * (2/5)
    c_height = height * (7/13)
    
    t.penup()
    t.goto(-152, 108)
    t.pendown()
    t.color(color)
    t.begin_fill()

    for i in range(2):
        t.forward(c_width)
        t.right(90)
        t.forward(c_height)
        t.right(90)
    t.end_fill()
    
## stars
def draw_stars():
    size = 8
    color = "#FFFFFF"
    
    t.color(color)
    t.begin_fill()
    
    for i in range(5):
        t.forward(size)
        t.right(144)
    t.end_fill()
    
def six_stars():
    x1 = - 145
    y1 = 95
    # loop for drawing multiple rows of stars
    for row in range(5):
        for star in range(6):
            t.penup()
            t.goto(x1, y1)
            t.pendown()
            draw_stars()
            x1 = x1 + 25
        # reset x1 to the starting position
        x1 = - 145
        # move down to the next row
        # modify this as you wish to set the distance between rows.
        y1 = y1 - 20
def five_stars():
    x2 = - 132
    y2 = 85
    # loop for drawing multiple rows of stars
    for row in range(4):
        for star in range(5):
            t.penup()
            t.goto(x2, y2)
            t.pendown()
            draw_stars()
            x2 = x2 + 25
        # reset x1 to the starting position
        x2 = - 132
        # move down to the next row
        # modify this as you wish to set the distance between rows.
        y2 = y2 - 20

draw_stripes()
draw_canton()
six_stars()
five_stars()
t.hideturtle
```
