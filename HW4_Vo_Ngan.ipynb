```python
#question 5
import matplotlib.pyplot as plt 
import numpy as np 

marks={'Andy':88,'Amy':66,'James':90,'Jules':77}
for student, grade in marks.items():
    print(f'{student} scores a grade of {grade}.')
```

```python
x=np.array(['Andy','Amy','James','Jules'])
y=np.array([88,66,90,77])

plt.barh(x,y)
plt.title("Student's Grade")
plt.show()
```

```python
y=np.array([88,66,90,77])
mylabels=['Andy','Amy','James','Jules']
plt.pie(y, labels=mylabels)
plt.title("Student's Grade")
plt.legend(bbox_to_anchor=(1.05,1.0),loc='upper left')
plt.show()
```

```python
#question 6
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5,6])
y=np.array([6,2,4,4,9,2])

plt.subplot(2,3,1)
plt.plot(x,y,x,y**2)
plt.title('Multiple Lines')

plt.subplot(2,3,2)
plt.bar(x,y)
plt.title('Bar')

plt.subplot(2,3,3)
mylabels=['A','B','C','D','E','F']
plt.pie(y, labels=mylabels)
plt.title('Pie')

plt.subplot(2,3,4)
plt.plot(x,y)
plt.grid()
plt.title('Grid')

plt.subplot(2,3,5)
plt.scatter(x,y)
plt.title('Scatterplot')

plt.subplot(2,3,6)
plt.hist(y)
plt.title('Histogram')

plt.show()
```
