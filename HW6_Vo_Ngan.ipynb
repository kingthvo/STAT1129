```r
max_salary<-max(df[,3])
min_salary<-min(df[,3])
med_salary<-median(df[,3])
```

```r
mylabel<-c("max salary","min salary","median salary")
colors<-c("blue","red","yellow")
x<-c(max_salary,min_salary,med_salary)
pie(x,label=mylabel,main="Salary Pie Chart",col=colors)
```
