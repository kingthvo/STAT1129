```python
import pandas as pd
df=pd.read_csv("amazon-orders.csv")
df.head()
```

```python
df.shape
```

```python
df=df.fillna(0)
df.head()
```

```python
df["Total Charged"].head()
```

```python
df["Total Charged"]=df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Total Charged"].describe()
```

```python
df["Total Charged"].median()
```

```python
df["Total Charged"].sum()
```

```python
df["Tax Charged"]=df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
print(df["Tax Charged"].sum())
print(df["Tax Charged"].sum()/df["Total Charged"].sum())
```

```python
df["Subtotal"]=df["Subtotal"].str.replace('$','').astype(float)
df.head()
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df.head()
```

```python
import matplotlib.pyplot as plt
df.plot.bar(x="Order Date",y="Total Charged",rot=90, figsize=(20,10))
```

```python
daily_orders=df.groupby("Order Date").sum()["Total Charged"]
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(20,10))
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
monthly_total=df.groupby(df['Order Date'].dt.strftime('%B'))['Total Charged'].sum()
monthly_total
```

```python
monthly_total.plot.bar(figsize=(20,10))
```

```python
monthly_mean=df.groupby(df['Order Date'].dt.strftime('%B'))['Total Charged'].mean()
monthly_mean
```

```python
monthly_mean.plot.bar(figsize=(20,10))
```

```python
monthly_max=df.groupby(df['Order Date'].dt.strftime('%B'))['Total Charged'].max()
monthly_max
```

```python
monthly_min=df.groupby(df['Order Date'].dt.strftime('%B'))['Total Charged'].min()
monthly_min
```

```python
monthly_median=df.groupby(df['Order Date'].dt.strftime('%B'))['Total Charged'].median()
monthly_median
```

```python
monthly_std=df.groupby(df['Order Date'].dt.strftime('%B'))['Total Charged'].std()
monthly_std
```

```python
monthly_summary=df.groupby(df['Order Date'].dt.strftime('%B'))['Total Charged'].describe()
monthly_summary
```
